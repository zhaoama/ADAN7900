```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy as sp
data = pd.read_csv('cities_data.clean.2.csv')
print(data.head())
```

```python
data['race_category'] = data[['prop_white', 'prop_black', 'pop.indian', 'pop.asian', 'pop.other']].idxmax(axis=1)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='high.school',
    y='total.volume',
    hue='bachelor',
    style='race_category',
    size='bachelor',
    data=data,
    palette='coolwarm',
    sizes=(50, 200),
    alpha=0.6
)
plt.title("Alcohol Consumption vs Education Level by White Population Proportion")
plt.xlabel("High School Completion (%)")
plt.ylabel("Alcohol Consumption (Volume)")
plt.legend(title="Education & Racial Group")
plt.show()
```
